テオヤンセンという人が考えた、リンク機構により擬似直線を作り出す機構です。テオヤンセン機構またはビースト機構と呼ばれます。
直線に動き、弧を描くように戻る動きを繰り返します。
テオヤンセン本人はこの機構を用いて風を受けて歩行するロボットを作りました。
テオヤンセン機構が作り出す動きは弧を描く部分が機構側に来るため、テオヤンセン機構を複数組み合わせるだけでロボットなどに歩行させることができます。
また、見てわかるように、このテオヤンセン機構には多くのリンクが使われています。
このリンクそれぞれの長さを調整することで、擬似直線や円弧っぽい挙動をカスタムすることができます。

```python
import piplite
await piplite.install(['ipycanvas==0.10.2', 'ipywidgets==7.7.1', 'ipyevents'])

import sys
if '.' not in sys.path:
    sys.path.append('.')
    
import scipy, sympy
from linkage import JansenLinkage    
```

```python
links = JansenLinkage(
    lengths = dict(
        a=38, b=41.5, c=39.3, d=40.1, e=55.8, f=39.4, 
        g=36.7, h=65.7, i=49.0, j=50, k=61.9, l=7.8, m=15.0)
)
links.solve_positions(30)
links.make_gui()
links.gui
```
